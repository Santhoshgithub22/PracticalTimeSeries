# Time Series Analysis

### Motivating Example: CO2 Emmision

- Plot time series in R
- Fit a linear model on the data

Linear models are given by
$$y_i = \beta_0 + \beta_1x_i + \epsilon_i$$ 
$\epsilon_i$ are the residuals    
**Assumptions for Linear Model:**
- $\epsilon_i$ are normally distributed with zero mean
- residuals have constant variance
- residuals are independent

In order to make inferences, we need to make distributional assumptions on $\epsilon_i$. 

````r
#The following two lines help to resize the plot
library(repr)
options(repr.plot.width=6, repr.plot.height=4)
plot(co2, main='Atmospheric Concentration of CO2')
````

````r
#put the command in paranthesis to force the output on screen
co2.lm=lm(co2~time(co2))
co2.lm
````

````r
plot(co2, main='Atmospheric Concentration of CO2')
abline(co2.lm)
````

## Examine the Residuals 

It' important to ensure that the residuals satisfy the assuptions we make about the model.

````r
resid.lm = resid(co2.lm)
attach(mtcars)
par(mfrow=c(1,2))
hist(resid.lm, prob=TRUE, main='Residuals')
lines(density(resid.lm))
qqnorm(resid.lm)
qqline(resid.lm)
````

The residuals appear to be close to **Normal Distribution**. Another to check normality is via **Q-Q Plot** which is a plot of theoretical quantile vs sample quantile. For normally distributed data, the plot should follow a straight line. Here we see small deviations at the bottom and top of the plot.

## CO2 Residuals vs Time

````r
options(repr.plot.width=8, repr.plot.height=4)
par(mfrow=c(1,2))
plot(resid.lm~time(co2), main='Residuals')
plot(resid.lm~time(co2), xlim=c(1960,1963), main='Zoomed Residuals')
````

The residuals don't appear to have zero mean. 

## Gosset Data

**Description**  
Data which show the effect of two soporific drugs (increase in hours of sleep compared to control) on 10 patients.   
**Format**   
A data frame with 20 observations on 3 variables.   

[, 1]	extra	numeric	increase in hours of sleep   
[, 2]	group	factor	drug given    
[, 3]	ID	factor	patient ID    

```R
#use the following to view infos on the data
help(sleep)
```

````r
plot(extra~group, data=sleep, main='Gosset Data')
````

- Looks like the median sleep for drug 2 is higher than that of drug 1
- Is the difference significant? **We need a statistical test**

````r
head(sleep)
````

$$t = \frac{(\bar{x_1}-\bar{x_2})-0}{\sigma/\sqrt{N}} = \frac{1.58-0}{1.229995483/\sqrt{10}} = -4.0621$$
$\sigma$ is the standard deviation of the difference between the two group

````r
attach(sleep)
extra.1 = extra[group==1]
extra.2 = extra[group==2]
t.test(extra.1, extra.2, paired=TRUE, alternative='two.sided')
````

- p-value is less than 0.05 so we reject the NULL
- Also, the 95 confidence interval does not contain 0, so we reject the NULL
